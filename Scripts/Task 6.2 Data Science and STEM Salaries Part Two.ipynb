```python
# Importing libraries and dataset
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

## Correlation Matrix Heatmap

```python
# Creating a correlation matrix
df_sal.corr()
```

```python
# Creating a subset to remove city_id from data (we do not need for heatmap purposes but do not want to drop completely)
df_sal2 = df_sal[['total_yearly_compensation', 'years_of_experience', 'years_at_company', 'base_salary', 'bonus', 'Masters_Degree',
                 'Bachelors_Degree', 'Doctorate_Degree', 'High_School', 'Some_College', 'Race_Asian', 'Race_White', 'Race_Two_Or_More',
                 'Race_Black', 'Race_Hispanic']]
```

```python
# Rechecking to see if city_id was removed
df_sal2.corr()
```

```python
# Creating a correlation heatmap using matplotlib
plt.matshow(df_sal2.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(14, 14)) 
plt.matshow(df_sal2.corr(), fignum=f.number)
plt.xticks(range(df_sal2.shape[1]), df_sal2.columns, fontsize=14, rotation=60) 
plt.yticks(range(df_sal2.shape[1]), df_sal2.columns, fontsize=14) # y axis labels
cb = plt.colorbar() 
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=14) 
```

```python
# Creating a heatmap using seaborn
f,ax = plt.subplots(figsize=(12,12))
corr = sns.heatmap(df_sal2.corr(), annot = True, ax = ax) 
```

## Creating a Scatterplot 

```python
# Creating a scatterplot for base_salary and total_yearly_compensation columns
sns.lmplot(x = 'base_salary', y = 'total_yearly_compensation', data = df_sal2)
```

#### This is a positive relationship. There are some outliers (which may be typos in the original data), and some of the data shows that some base salaries start at 0. While this may be true for some jobs in this field, this could also mean that there wasn't any data entered or available, which is throwing off our results. 

```python
# Creating a scatterplot for years_at_company and years_of_experience columns
sns.lmplot(x = 'years_of_experience', y = 'years_at_company', data = df_sal2)
```

#### There is a positive relationship. There are a few outliers, but for the most part, we can assume that the majority of experience comes from the company the employee is currently working.

## Creating a Categorical Plot

```python
# Creating a categorical plot that splits total_yearly_compensation into categories
sns.histplot(df_sal2['total_yearly_compensation'], bins = 20, kde = True);
```

```python
df_sal2.loc[df_sal2['total_yearly_compensation'] < 20000, 'Total Compensation'] = 'Low End'
```

```python
df_sal2.loc[(df_sal2['total_yearly_compensation'] > 200000) & (df_sal2['total_yearly_compensation'] < 500000), 'Total Compensation'] = 'Middle'
```

```python
df_sal2.loc[df_sal2['total_yearly_compensation'] >= 500000, 'Total Compensation'] = 'High End'
```

```python
df_sal2['Total Compensation'].value_counts(dropna = False)
```

```python
# Creating a categorical plot in seaborn using the categories created above
sns.set(style="ticks")
g = sns.catplot(x="base_salary", y="total_yearly_compensation", hue="Total Compensation", data=df_sal2)
```
